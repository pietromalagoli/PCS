Firstly, I try to reproduce the mean-field approach. The resulting system of ODEs is the following:
$$
    \frac{\partial x}{\partial t} = \alpha x - \gamma xy , \;\;\;
    \frac{\partial y}{\partial t} = \lambda + \nu xy - \sigma y
$$
where $x$ is the pathogen concentration and $y$ the lymphocytes'. \
The parameters represent:
- $\alpha$ : proliferation rate of the pathogen; 
- $\gamma$ : destruction rate of the pathogen by the lymphocytes; 
- $\lambda$ : birth rate of the lymphocyte; 
- $\nu$ : duplication rate of the lymphocyte when encountering the pathogen; 
- $\sigma$ : death rate of the lymphocyte.

The time it takes for $y$ to reach $\tilde{y}$, which is the value of $y$ for which $x=1$, is approximately:
$$
\Delta t_1 = \frac{1}{\gamma}\ln\left(\frac{\tilde{y}-\frac{\lambda}{\sigma}}{\frac{\alpha}{\gamma}-\frac{\lambda}{\sigma}}\right)
$$
while, the time it takes for $x$ to increase from 1 to a larger number $k$ is:
$$
\Delta t_2 = \frac{1}{\alpha}\ln k
$$

    Y \xrightarrow{\sigma} \emptyset .
\end{gather}
$$

Now diffusion is added to the stochastic model. In particular, we have an isotropic discrete diffusion regulated by two coefficients, $D_{x}$ and $D_{y}$, for the two populations, respectively pathogens ($x$) and immune system ($y$). The diffusive process is carried out as a Poisson number with mean, e.g. for the pathogen population, $\lambda=D_{x}x$ (where $x$ is the population of the starting site), subtracted from the starting site and added to its neighbours.

```python
from matplotlib.animation import PillowWriter

def lattice1(time_steps:int, par:dict, N:int=100, rng=None,
             animate:bool=False, gif_path:str=None, capture_interval:int=1, fps:int=10):
    """
    Vectorized stochastic lattice. Uses a numpy Generator for faster poisson draws.

    New optional args:
    - animate: if True, capture occupancy (empty / non-empty) over time and optionally save as GIF.
    - gif_path: if provided and animate is True, the GIF will be written to this path.
    - capture_interval: sample every `capture_interval` timesteps to build frames (reduces memory).
    - fps: frames per second when writing the GIF.

    Backwards compatible: if animate is False, returns (X_stoc, Y_stoc) as before.
    If animate is True, returns (X_stoc, Y_stoc, gif_path) after saving (or None if gif_path is None).
    """
    if rng is None:
        rng = np.random.default_rng()
    alpha = par['alpha']        # pre-extract parameters (avoid repeated dict lookups)
    gamma = par['gamma']
    lam = par['lambda']
    nu = par['nu']
    sigma = par['sigma']
    Dx = par['Dx']
    Dy = par['Dy']

    lattice = np.zeros((N, 2), dtype=np.int64)      # integer lattice for counts
    lattice[:, 0] = 3
    if Dx and Dy == 0:      # keep original logic (as in the notebook)
        lattice[:, 1] = 1
    else:
        lattice[:, 1] = rng.poisson(lam=1)
    lattice_old = lattice.copy()
    X_stoc = np.zeros(time_steps, dtype=float)
    Y_stoc = np.zeros(time_steps, dtype=float)

    # Prepare animation buffers if requested
    if animate:
        # map 1D lattice to near-square 2D grid for visualization
        ncols = int(np.ceil(np.sqrt(N)))
        nrows = int(np.ceil(N / ncols))
        total_cells = nrows * ncols
        n_frames = int(np.ceil(time_steps / float(max(1, capture_interval))))
        frames = np.zeros((n_frames, nrows, ncols), dtype=np.uint8)
        frame_idx = 0

    for t in range(time_steps):
        prod = lattice_old[:, 0] * lattice_old[:, 1]            # elementwise xy
        # vectorized Poisson draws
        births_x = rng.poisson(alpha * lattice_old[:, 0])
        kills_x = rng.poisson(gamma * prod)
        births_y = rng.poisson(lam, size=N)
        dup_y = rng.poisson(nu * prod)
        deaths_y = rng.poisson(sigma * lattice_old[:, 1])
        # update lattice (vectorized)
        lattice[:, 0] = births_x - kills_x
        lattice[:, 1] = births_y + dup_y - deaths_y
        if Dx or Dy > 0:    # diffusion
            diffX = rng.poisson(np.sqrt(Dx*lattice_old[:,0]))
            diffY = rng.poisson(np.sqrt(Dy*lattice_old[:,1]))
            np.minimum(diffX,lattice_old[:,0],out=diffX)    # the maximum of individuals to diffuse out is the population of the site
            np.minimum(diffY,lattice_old[:,1],out=diffY)
            diffX = diffX // 2      # divide the # of individuals to diffuse from each site by the # of neighbours
            diffY = diffY // 2      # approximate down
            lattice[:,0] -= diffX   # take out the diffused individuals
            lattice[:,1] -= diffY
            lattice[:,0] += np.roll(diffX,+1) + np.roll(diffX,-1)  # add the diffused individuals
            lattice[:,1] += np.roll(diffY,+1) + np.roll(diffY,-1)  
        np.maximum(lattice, 0, out=lattice)         # clip negatives in-place (no negative populations)
        lattice_old = lattice.copy()           # prepare for next step (must copy to avoid aliasing)
        # mean
        X_stoc[t] = lattice[:, 0].mean()
        Y_stoc[t] = lattice[:, 1].mean()

        # capture occupancy frame if requested
        if animate and (t % capture_interval == 0):
            occupied = ((lattice[:, 0] > 0) | (lattice[:, 1] > 0)).astype(np.uint8)
            # pad to fill the grid
            if total_cells > N:
                padded = np.zeros(total_cells, dtype=np.uint8)
                padded[:N] = occupied
            else:
                padded = occupied[:total_cells]
            frames[frame_idx, :, :] = padded.reshape(nrows, ncols)
            frame_idx += 1

    # If animation requested, create and optionally save GIF (black/white: empty vs non-empty)
    if animate:
        try:
            import matplotlib.animation as animation
            pillow_writer_available = True
        except Exception:
            animation = None
            pillow_writer_available = False

        fig = plt.figure(figsize=(6, 6))
        ax = fig.add_subplot(111)
        ax.axis('off')
        im = ax.imshow(frames[0], cmap='Greys_r', vmin=0, vmax=1, interpolation='nearest')

        def _update(i):
            im.set_data(frames[i])
            return (im,)

        ani = None
        if animation is not None:
            ani = animation.FuncAnimation(fig, _update, frames=frame_idx, blit=True)
            if gif_path and pillow_writer_available and (PillowWriter is not None):
                writer = PillowWriter(fps=fps)
                ani.save(gif_path, writer=writer)
        plt.close(fig)
        # Return gif path as third output for convenience (None if not saved)
        return X_stoc, Y_stoc

    return X_stoc, Y_stoc
```

```python
def lattice2(time_steps:int, par:dict, N:int=100, rng=None,
             animate:bool=False, gif_path:str=None, capture_interval:int=1, fps:int=10):
    """
    Vectorized stochastic 2D lattice. Uses a numpy Generator for faster poisson draws.

    New optional args:
    - animate: if True, capture occupancy (empty / non-empty) over time and optionally save as GIF.
    - gif_path: if provided and animate is True, the GIF will be written to this path.
    - capture_interval: sample every `capture_interval` timesteps to build frames (reduces memory).
    - fps: frames per second when writing the GIF.

    Backwards compatible: if animate is False, returns (X_stoc, Y_stoc) as before.
    If animate is True, returns (X_stoc, Y_stoc) and will save the GIF if gif_path is provided.
    """
    if rng is None:
        rng = np.random.default_rng()
    alpha = par['alpha']        # pre-extract parameters (avoid repeated dict lookups)
    gamma = par['gamma']
    lam = par['lambda']
    nu = par['nu']
    sigma = par['sigma']
    Dx = par['Dx']
    Dy = par['Dy']

    lattice = np.zeros((N, N, 2), dtype=np.int64)      # integer lattice for counts
    lattice[:, :, 0] = 3
    if Dx and Dy == 0:      # no diffusion (kept consistent with earlier cells)
        lattice[:, :, 1] = 1
    else:
        lattice[:, :, 1] = rng.poisson(lam=1, size=(N, N))
    lattice_old = lattice.copy()
    X_stoc = np.zeros(time_steps, dtype=float)
    Y_stoc = np.zeros(time_steps, dtype=float)

    # Prepare animation buffers if requested
    if animate:
        n_frames = int(np.ceil(time_steps / float(max(1, capture_interval))))
        frames = np.zeros((n_frames, N, N), dtype=np.uint8)
        frame_idx = 0

    for t in range(time_steps):
        prod = lattice_old[:, :, 0] * lattice_old[:, :, 1]            # elementwise xy
        # vectorized Poisson draws
        births_x = rng.poisson(alpha * lattice_old[:, :, 0])
        # kills_x and dup_y might be zero if prod is zero-sized but here prod is (N,N)
        kills_x = rng.poisson(gamma * prod)
        births_y = rng.poisson(lam, size=(N, N))
        dup_y = rng.poisson(nu * prod)
        deaths_y = rng.poisson(sigma * lattice_old[:, :, 1])
        # update lattice (vectorized)
        lattice[:, :, 0] = births_x - kills_x
        lattice[:, :, 1] = births_y + dup_y - deaths_y
        if Dx or Dy > 0:    # diffusion (kept consistent with lattice1)
            diffX = rng.poisson(np.sqrt(Dx * lattice_old[:, :, 0]))
            diffY = rng.poisson(np.sqrt(Dy * lattice_old[:, :, 1]))
            np.minimum(diffX, lattice_old[:, :, 0], out=diffX)    # the maximum of individuals to diffuse out is the population of the site
            np.minimum(diffY, lattice_old[:, :, 1], out=diffY)
            diffX = diffX // 4      # divide the # of individuals to diffuse from each site by the # of neighbours (4 in 2D)
            diffY = diffY // 4      # approximate down
            lattice[:, :, 0] -= diffX   # take out the diffused individuals
            lattice[:, :, 1] -= diffY
            # add the diffused individuals from the four neighbours
            lattice[:, :, 0] += np.roll(diffX, +1, axis=0) + np.roll(diffX, -1, axis=0)
            lattice[:, :, 0] += np.roll(diffX, +1, axis=1) + np.roll(diffX, -1, axis=1)
            lattice[:, :, 1] += np.roll(diffY, +1, axis=0) + np.roll(diffY, -1, axis=0)
            lattice[:, :, 1] += np.roll(diffY, +1, axis=1) + np.roll(diffY, -1, axis=1)
        np.maximum(lattice, 0, out=lattice)         # clip negatives in-place (no negative populations)
        lattice_old = lattice.copy()           # prepare for next step (must copy to avoid aliasing)
        X_stoc[t] = lattice[:, :, 0].mean()    # mean
        Y_stoc[t] = lattice[:, :, 1].mean()

        # capture occupancy frame if requested
        if animate and (t % capture_interval == 0):
            occupied = (lattice[:, :, 0] > 0).astype(np.uint8)
            frames[frame_idx, :, :] = occupied
            frame_idx += 1

    # If animation requested, create and optionally save GIF (black/white: empty vs non-empty)
    from matplotlib.patches import Rectangle
    if animate:
        try:
            import matplotlib.animation as animation
            pillow_writer_available = True
        except Exception:
            animation = None
            Rectangle = None
            pillow_writer_available = False

        fig = plt.figure(figsize=(6, 6))
        ax = fig.add_subplot(111)
        ax.axis('off')
        im = ax.imshow(frames[0], cmap='Greys_r', vmin=0, vmax=1, interpolation='nearest')

        # draw lattice borders and light grid lines between cells
        # ticks at half-integers align with pixel boundaries for imshow
        ax.set_xticks(np.arange(-0.5, N, 1))
        ax.set_yticks(np.arange(-0.5, N, 1))
        ax.set_xticklabels([])
        ax.set_yticklabels([])
        ax.grid(which='major', color='black', linestyle='-', linewidth=0.25, alpha=0.5)

        # outer rectangle border for clarity (stands out from the internal grid)
        if Rectangle is not None:
            rect = Rectangle((-0.5, -0.5), N, N, fill=False, edgecolor='black', linewidth=1.0, alpha=0.9)
            ax.add_patch(rect)

        # Add a timestamp text in axes coordinates (top-left) with a semi-opaque box for readability
        timestamp_text = ax.text(
            0.01, 0.97, "", transform=ax.transAxes,
            color='red', fontsize=10, va='top', ha='left',
            bbox=dict(facecolor='black', alpha=0.3, pad=2)
        )

        def _update(i):
            im.set_data(frames[i])
            t_step = i * capture_interval
            timestamp_text.set_text(f"t = {t_step}")
            return (im, timestamp_text)

        ani = None
        if animation is not None:
            # frame_idx is the actual number of frames filled during the simulation
            frames_to_use = frame_idx if frame_idx > 0 else 1
            ani = animation.FuncAnimation(fig, _update, frames=frames_to_use, blit=True)
            if gif_path and pillow_writer_available and (PillowWriter is not None):
                writer = PillowWriter(fps=fps)
                ani.save(gif_path, writer=writer)
        plt.close(fig)
        return X_stoc, Y_stoc

    return X_stoc, Y_stoc
```
